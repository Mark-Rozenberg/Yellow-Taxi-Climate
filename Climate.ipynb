```python
# URL for the PSV file
for station_id in station_ids:
    # url_psv = f'https://www.ncei.noaa.gov/oa/global-historical-climatology-network/hourly/access/by-station/GHCNh_{station_id}_por.psv'
    url_psv = f'https://www.ncei.noaa.gov/oa/global-historical-climatology-network/hourly/access/by-year/2023/psv/GHCNh_{station_id}_2023.psv'
    # Send a GET request to the URL
    response_psv = requests.get(url_psv, headers=headers)

    # Save the content to a PSV file
    with open(f'Data/Climate/2023/{station_id}.psv', 'wb') as file_psv:
        file_psv.write(response_psv.content)

    print(f"PSV file for station {station_id} downloaded successfully.")
```

```python
import os

# Directory containing the PSV files
climate_dir = 'Data/Climate/2023'

# List to hold DataFrames
# df_list = []

# Loop through all files in the directory
for filename in os.listdir(climate_dir):
    if filename.endswith('.psv'):
        file_path = os.path.join(climate_dir, filename)
        df_temp = pd.read_csv(file_path, sep='|')
        df_list.append(df_temp)

# Combine all DataFrames into one
df = pd.concat(df_list, ignore_index=True)
```

```python
df = df.loc[:, ~df.columns.str.endswith('_Measurement_Code')]
df = df.loc[:, ~df.columns.str.endswith('_Quality_Code')]
df = df.loc[:, ~df.columns.str.endswith('_Report_Type')]
df = df.loc[:, ~df.columns.str.endswith('_Source_Code')]
df = df.loc[:, ~df.columns.str.endswith('_Source_Station_ID')]
df = df.loc[:, ~df.columns.str.startswith('<?xml')]
```

```python
# Convert DATE column to datetime
df['DATE'] = pd.to_datetime(df['DATE'])

# Group by Station_ID and aggregate the number of unique dates, min date, and max date
result = df.groupby('Station_ID')['DATE'].agg(
    unique_dates='nunique',
    min_date='min',
    max_date='max'
).reset_index()

result
```
